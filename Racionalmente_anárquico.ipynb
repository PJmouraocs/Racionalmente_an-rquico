```python
import matplotlib.pyplot as plt
import numpy as np

# Definir os intervalos de x que queremos plotar
theta = np.linspace(0, 2*np.pi, 100)
x = 2 + np.cos(theta)
y = 2 + np.sin(theta)

# Definir as funções h, A e j
h = lambda x: 3*x - 3
A = lambda x: -3*x + 9
j = lambda x: 0.2*x + 1.4

# Definir um novo intervalo de valores de x para j
x_j = np.linspace(-1, 5, 1000)

# Plotar as funções
plt.plot(x, y, label="f(x) e g(x)")
plt.plot(x, h(x), label="h(x)")
plt.plot(x, A(x), label="A(x)")
plt.plot(x_j, j(x_j), label="j(x)")

# Configurar o gráfico
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Mostrar o gráfico
plt.show()
```
